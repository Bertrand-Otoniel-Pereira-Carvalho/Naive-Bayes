```python
import pandas as pd #tratar tabelas
import numpy as np #tratar matrizes e listas
from sklearn.model_selection import train_test_split #Machine Learning
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report
#from yellowbrick.classifier import ConfusionMatrix #Vizualização de desempenho de modelos de Machine Learning
```

```python
base = pd.read_csv('insurance.csv',keep_default_na = False) # Lendo o arquivo csv
```

```python
base.head()
```

```python
base = base.drop(columns =["Unnamed: 0"]) # retirando uma coluna
base.head() # Exibe apenas as 5 primeiras linhas
```

```python
base.shape # qtd de linhas e colunas
```

```python
print(base.isnull().sum())
```

```python
y = base.iloc[:,7].values # y = coluna 7

x = base.drop(base.columns[7],axis = 1).values # x = a todas as colunas menos a 7
```

```python
labelencoder = LabelEncoder()
for i in range(x.shape[1]):
    if x[:,i].dtype == 'object':
        x[:,i] = labelencoder.fit_transform(x[:,i])

```

```python
x_treinamento,x_test,y_treinamento,y_test = train_test_split(x,y,test_size=0.3,random_state=12) # Retornar aqui, entender conceitos
```

```python
modelo = GaussianNB()
modelo.fit(x_treinamento,y_treinamento)
```

```python
previsoes = modelo.predict(x_test)
```

```python
previsoes
```

```python
accuracy = accuracy_score(y_test,previsoes)
precision = precision_score(y_test,previsoes,average="weighted")
recall = recall_score(y_test,previsoes,average="weighted")
f1 = f1_score(y_test,previsoes,average="weighted")
print(f"Acurácia:{accuracy}, Precisão {precision}, Recall: {recall},F1 {f1}")
```

```python
report = classification_report(y_test,previsoes)
print(report)
```
